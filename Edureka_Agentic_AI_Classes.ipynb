```python
#Multi agent system
from datetime import datetime
import os
import google.generativeai as genai
GOOGLE_API_KEY="AQ.Ab8RN6LlikOvgf6zgQvLkxmdPvvZ_r-POoaf_S_kdjZ0nT3WFg"
genai.configure(api_key=GOOGLE_API_KEY)
model=genai.GenerativeModel("gemini-pro-latest")

#Agent 1 :Symptom Agent
def symptom_agent(input_text):
  print('Symptom Agent')
  chat=model.start_chat()
  response=model.generate_content(f'Classify the severity of the symptoms and list possible treatment:{input_text}')
  print(response.text)
  return response.text

#Agent2:Schduelar Agent
def schedular_agent(symptom_severity_summary):
  print('Schedular Agent')
  response=model.generate_content(f'Based on this summary,suggest next steps for patient{symptom_severity_summary}')
  print(response.text)
  return response.text


#Follow up Agent
def follow_up_agent(follow_up_question):
  print('follow up Agent')
  response=model.generate_content(f'What follow up questions should a dr or a nurse as a patient condition better {follow_up_question}')
  print(response.text)
  return response.text

# Agent 4: Log this Agent
def logger_agent(name,pid,age,input_text,symptom_severity_summary,follow_up_question):
 print('\n[Logger Agent] Recording Information')
 log_entry=f"\nTime:{datetime.now()}\nName:{name}\nPatient ID:{pid}\nAge:{age}\nInput Text:{input_text}\nSymptom Severity Summary:{symptom_severity_summary}\nFollow Up Question:{follow_up_question}"
 with open("patient_log.txt","a") as f:
  f.write(log_entry)


#Get User Id and other information
name=input("Enger the patient Name:")
pid=input("Enter the patient Id :")
age=input("Enter the patients Age:")
user_input = input("Please describe your health symptoms")

#Run Agents
summary=symptom_agent(user_input)

#Follow up Agent
next_steps=schedular_agent(summary)

followups=follow_up_agent(next_steps)

logger_agent(name,pid,age,user_input,summary,followups)
```
