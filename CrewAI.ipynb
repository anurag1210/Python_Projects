```python
!pip install -U crewai litellm

import nest_asyncio
nest_asyncio.apply()

print('installation complete')

#Simple agent

from crewai import Agent, Task, Crew, LLM
import os
from getpass import getpass

gemini_api_key = getpass("Gemini API Key :")
os.environ['GEMINI_API_KEY'] = gemini_api_key

llm = LLM(
    model="gemini/gemini-2.5-flash",
    temperature=0.7,
    api_key=gemini_api_key
)

#Get Groq API key

print('Enter your groq API key')
print('get one free at https://console.groq.com/keys')


print("LLM initialized")

#Create a simple research agent

researcher=Agent(
    role="Research Analyst",
    goal="Research some topic and provide clear, concise insights",
    backstory="You are a skilled researched who provides accurate information",
    llm=llm,
    verbose=True
)
print("researcher agent initialized")

##Create a simple Task

research_task=Task(

     description="""Research the topic:{topic}

     provide brief overview in 3-4 sentences.
     focus on the most important point.""",
          agent=researcher,
          expected_output="A concise 3-4 sentences of the topic",
          verbose=True
)

crew=Crew(
    agents=[researcher],
    tasks=[research_task]
)

###Execute the task
print("\n Starting research")
print("="*50)


result = await crew.kickoff_async(inputs={
    "topic": "What is AI and how does it work"
})
print(result)
```
